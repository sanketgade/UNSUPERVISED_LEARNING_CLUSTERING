```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++',random_state=0)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
pd.concat([data, pd.DataFrame(y_kmeans)], axis=1)
```

```python
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1], s=100, c='red', label='cluster1')
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1], s=100, c='pink', label='cluster1')
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', 
                             linkage='ward')
```

```python
y_hc = hc.fit_predict(x)
```

```python
y_hc
```

```python
plt.scatter(x[y_hc==0,0], x[y_hc==0,1], s=100, c='red', label='cluster1')
plt.scatter(x[y_hc==1,0], x[y_hc==1,1], s=100, c='pink', label='cluster1')
```
